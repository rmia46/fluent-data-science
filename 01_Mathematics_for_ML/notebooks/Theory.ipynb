# Module 01: Mathematics for Machine Learning & Data Science

Mathematics is the bedrock of Machine Learning. This module bridges core mathematical concepts with their practical implementation in Python, specifically focusing on how these concepts power ML algorithms.

## 1. Linear Algebra: The Language of Data

### Why it matters in ML:
- **Data Representation:** Every dataset is essentially a matrix where rows are observations and columns are features.
- **Neural Networks:** Weights and biases are stored as matrices, and forward/backward passes are matrix operations.
- **Dimensionality Reduction:** Techniques like PCA (Principal Component Analysis) use SVD and Eigenvalues to find the most important directions in data.

### Core Concepts:
- **Vectors & Matrices:** The basic building blocks.
- **Dot Product:** Measures similarity and is the basis of the "weighted sum" in neurons.
- **Matrix Multiplication:** How data flows through layers.
- **Eigenvalues & Eigenvectors:** Understanding transformations and variance.

### Python Implementation:

```python
import numpy as np

# Representing a small dataset (3 samples, 2 features)
X = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

# Weight matrix (2 features to 1 output)
W = np.array([0.5, -0.5])

# Dot product: predicting output for all samples
y_pred = X.dot(W)
print(f"Predictions: {y_pred}")
```

## 2. Calculus: The Engine of Optimization

### Why it matters in ML:
- **Gradient Descent:** We find the minimum of a loss function by taking the derivative (gradient) and moving in the opposite direction.
- **Backpropagation:** Uses the **Chain Rule** to distribute errors back through a network to update weights.

### Core Concepts:
- **Derivatives:** Rate of change.
- **Partial Derivatives:** Rate of change with respect to one variable (holding others constant).
- **Gradients:** Vectors of partial derivatives pointing to the steepest ascent.

### Python Implementation:

```python
# Simple Gradient Descent to minimize f(x) = x^2
def f(x): return x**2
def df(x): return 2*x

x = 10.0  # Initial guess
lr = 0.1  # Learning rate

for i in range(10):
    gradient = df(x)
    x = x - lr * gradient
    print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")
```

## 3. Probability & Statistics: Quantifying Uncertainty

### Why it matters in ML:
- **Distributions:** Most models assume something about how data is distributed (e.g., Gaussian/Normal distribution for Linear Regression).
- **Bayes' Theorem:** Foundation for Naive Bayes and understanding conditional dependencies.
- **Evaluation:** Hypothesis testing and confidence intervals help us know if a model improvement is statistically significant.

### Core Concepts:
- **Mean, Median, Variance, Std Dev:** Descriptive summaries.
- **Normal Distribution:** The "Bell Curve" often found in natural data.
- **Z-Scores:** Standardizing data (scaling) for models like SVM or KNN.

### Python Implementation:

```python
import matplotlib.pyplot as plt

# Generating Normal Data
data = np.random.normal(loc=50, scale=10, size=1000)

print(f"Mean: {np.mean(data):.2f}")
print(f"Std Dev: {np.std(data):.2f}")

# Scaling (Z-score normalization)
scaled_data = (data - np.mean(data)) / np.std(data)
print(f"Scaled Mean: {np.mean(scaled_data):.2f} (Expected 0)")
print(f"Scaled Std Dev: {np.std(scaled_data):.2f} (Expected 1)")
```
