# Module 03 Theory: Feature Engineering I

Feature Engineering is the art of creating new predictive signals. This notebook demonstrates the core transformations used in competitive machine learning.

## 1. Categorical Encoding
Transforming text into numbers based on cardinality and order.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'City': ['Dhaka', 'Chittagong', 'Dhaka', 'Sylhet', 'Dhaka']})

# 1. One-Hot Encoding (Low Cardinality)
ohe = pd.get_dummies(df['City'], prefix='City')
print("One-Hot Encoding:\n", ohe)

# 2. Ordinal Encoding (Ranked Categories)
df_rank = pd.DataFrame({'Rank': ['Low', 'High', 'Medium']})
rank_map = {'Low': 0, 'Medium': 1, 'High': 2}
df_rank['Rank_Encoded'] = df_rank['Rank'].map(rank_map)
print("\nOrdinal Encoding:\n", df_rank)

# 3. Frequency Encoding (High Cardinality)
freq = df['City'].value_counts(normalize=True)
df['City_Freq'] = df['City'].map(freq)
print("\nFrequency Encoding:\n", df)
```

## 2. Numerical Transformations
Scaling data and taming skewed distributions.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

df_num = pd.DataFrame({'Age': [20, 30, 40, 50, 60], 'Income': [50000, 60000, 70000, 1000000, 55000]})

# 1. Standard Scaling (Mean=0, Std=1)
scaler = StandardScaler()
df_num['Age_Scaled'] = scaler.fit_transform(df_num[['Age']])

# 2. Log Transformation (Taming Right-Skewed Data)
df_num['Income_Log'] = np.log1p(df_num['Income'])
print("Numerical Transformations:\n", df_num)
```

## 3. Interaction Features & Binning
Creating multi-column relationships and simplifying noise.

```python
# 1. Interaction Feature (Ratio)
df_int = pd.DataFrame({'Calories': [500, 1000], 'Grams': [100, 250]})
df_int['Cal_Per_Gram'] = df_int['Calories'] / df_int['Grams']

# 2. Binning (Categorical Buckets)
df_age = pd.DataFrame({'Age': [10, 25, 45, 70]})
df_age['Age_Bin'] = pd.cut(df_age['Age'], bins=[0, 18, 35, 60, 100], labels=['Minor', 'Young', 'Middle', 'Senior'])
print("Interactions & Binning:\n", df_int, "\n\n", df_age)
```
