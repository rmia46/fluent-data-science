# Module 04 Theory: Advanced Golden Features & Validation

In competitive data science, raw features are rarely enough. This notebook demonstrates the practical implementation of 'Golden Features'---transformations that capture hidden relationships in the data.

## 1. Smoothed Target Encoding
How to handle high-cardinality categories using the target signal without overfitting.

```python
import pandas as pd
import numpy as np

df_sample = pd.DataFrame({
    'Cat': ['A', 'A', 'B', 'C', 'A', 'B'],
    'Target': [1, 0, 1, 0, 1, 0]
})

m = 2 # Smoothing
global_mean = df_sample['Target'].mean()
agg = df_sample.groupby('Cat')['Target'].agg(['count', 'mean'])

smoothed = (agg['mean'] * agg['count'] + global_mean * m) / (agg['count'] + m)
df_sample['Cat_Encoded'] = df_sample['Cat'].map(smoothed)
print("Smoothed Target Encoding:\n", df_sample)
```

## 2. Group-Level Contextual Transforms
Comparing individual data points against peer group statistics.

```python
df_peer = pd.DataFrame({
    'Group': ['X', 'X', 'Y', 'Y'],
    'Val': [100, 200, 10, 20]
})

df_peer['Group_Mean'] = df_peer.groupby('Group')['Val'].transform('mean')
df_peer['Relative_Val'] = df_peer['Val'] / df_peer['Group_Mean']
print("Contextual Transforms:\n", df_peer)
```

## 3. Automated Feature Selection
Ranking features by their predictive contribution.

```python
import lightgbm as lgb
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=100, n_features=5, n_informative=2, random_state=42)
model = lgb.LGBMClassifier(verbose=-1).fit(X, y)

importances = pd.Series(model.feature_importances_, index=[f'feat_{i}' for i in range(5)])
print("Feature Importances:\n", importances.sort_values(ascending=False))
```
