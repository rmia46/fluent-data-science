## 1. Vectorization vs. Iteration
Always prefer vectorized operations over `for` loops or `.iterrows()`.

```python
import pandas as pd
import numpy as np
import time

# Create a large dummy dataset
df_large = pd.DataFrame({
    'A': np.random.rand(1000000),
    'B': np.random.rand(1000000)
})

# 1. SLOW: Iteration
start = time.time()
res_loop = []
for i, row in df_large.head(1000).iterrows(): # Only doing 1000 because it's so slow!
    res_loop.append(row['A'] + row['B'])
print(f"Iteration (1k rows): {time.time() - start:.4f}s")

# 2. FAST: Vectorization
start = time.time()
df_large['C'] = df_large['A'] + df_large['B']
print(f"Vectorization (1M rows): {time.time() - start:.4f}s")
```

## 2. Advanced Filtering (.loc vs .iloc)
Use `.loc` for labels and `.iloc` for positions.

```python
df = pd.DataFrame({'Score': [10, 20, 30]}, index=['a', 'b', 'c'])

print("Label based (.loc['a']):", df.loc['a']['Score'])
print("Position based (.iloc[0]):", df.iloc[0]['Score'])

# Bitwise operators for complex filters
mask = (df['Score'] > 15) & (df['Score'] < 25)
print("\nFiltered Rows:\n", df[mask])
```

## 3. Relational Operations (Groupby & Merge)
Summarizing data and linking tables.

```python
df_sales = pd.DataFrame({
    'User': ['A', 'B', 'A', 'C'],
    'Amount': [100, 200, 150, 300]
})

# Groupby Aggregation
summary = df_sales.groupby('User')['Amount'].agg(['mean', 'sum'])
print("User Summary:\n", summary)

# Merge (Left Join)
df_users = pd.DataFrame({'User': ['A', 'B', 'D'], 'Age': [25, 30, 22]})
merged = df_sales.merge(df_users, on='User', how='left')
print("\nMerged Data (Note how 'D' is lost and 'C' has NaN Age):\n", merged)
```

## 4. Cleaning & Indicators
Tracking missingness for better models.

```python
df_messy = pd.DataFrame({'Age': [25, np.nan, 30, 40, np.nan]})

# 1. Create indicator
df_messy['Age_was_missing'] = df_messy['Age'].isna().astype(int)

# 2. Impute with median
df_messy['Age'] = df_messy['Age'].fillna(df_messy['Age'].median())
print("Cleaned Data with Indicator:\n", df_messy)
```
